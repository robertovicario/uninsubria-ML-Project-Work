# Exploratory Data Analysis (EDA)

### Raw Data

```python
from sklearn.conftest import fetch_olivetti_faces

data = fetch_olivetti_faces()
images, labels = data.images, data.target
```

### Figure 1.1: Grid of sample images of all individuals inside the dataset.

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(4, 10)
axes = axes.flatten()

count_per_label = [0] * 40
index = 0

for i in range(len(images)):
    label = labels[i]
    if count_per_label[label] == 0:
        axes[index].imshow(images[i], cmap='gray')
        axes[index].axis('off')
        axes[index].set_title(label+1, fontsize=10)
        count_per_label[label] = 1
        index += 1
        
        if index >= len(axes):
            break

plt.tight_layout()
plt.show()
```
